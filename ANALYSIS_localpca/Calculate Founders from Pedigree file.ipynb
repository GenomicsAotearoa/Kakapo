```python
import pandas as pd
```

```python
family = pd.read_csv("Pedigree_22_March_2021.txt", sep="\s+", header=None)
family
```

```python
def is_founder(row):
    return row[2] == "-9" and row[3] == "-9"

family["founder"] = family.apply(lambda row: is_founder(row), axis=1)
family
```

```python
founders = family[family["founder"] == True]
founders.to_csv("founders.ped", sep="\t", index=False)
founders.head()
```
