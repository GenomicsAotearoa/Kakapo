```python
import pandas as pd
import numpy as np
```

```python
bin_size = 100000
data = pd.read_csv("Ho_He_groups.csv")
data.head()
```

```python
f_vs_o = data[["CHR", "POS", "He_Founders", "He_Gen3"]].copy()
f_vs_o["bin"] = list(f_vs_o["POS"] // bin_size)
f_vs_o = f_vs_o[["CHR", "bin", "He_Founders", "He_Gen3"]].copy()
f_vs_o['Change'] = f_vs_o['He_Gen3'] - f_vs_o['He_Founders']
f_vs_o = f_vs_o.groupby(['CHR', 'bin']).mean().reset_index().copy()
f_vs_o.head()

with open("he_f_vs_gen3.scatter", "w") as writer:
    for i,row in f_vs_o.iterrows():
        writer.write("\t".join(map(str, [row['CHR'], row['bin'] * bin_size, row['bin'] * bin_size + bin_size, row['Change']])))
        writer.write("\n")
```
